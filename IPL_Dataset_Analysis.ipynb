**Which season consisted of the highest no of matches ever played**

```python
fig_dims=(20,5)
fig, ax=plt.subplots(figsize=fig_dims)
sns.countplot(x='season',ax=ax,data=dataset)
plt.show()
```

From the above chart, we can say that the 2013 IPL Season consisted of maximum no of IPL Matches.

**What is the probability of winning if the toss is won by the team**

```python
probability_of_win = dataset['toss_winner']==dataset['winner']

probability_of_win.groupby(probability_of_win).size()
```

**Which is the most successful team with respect to this available data (Maximum no of matches won)**

```python
data1=dataset['winner'].value_counts()
sns.barplot(y=data1.index,x=data1)
```

From the above bar plot, we can say that Mumbai Indians is the most successful team

**Which Stedium has hosted maximum no of IPL Matches**

```python
dataset['venue'].value_counts()
```

```python
data2=dataset['venue'].value_counts()
sns.barplot(y=data2.index,x=data2)
```

From the above bar plot, we can say that M.Chinnaswamy Stadium has hosted the maximum IPL matches.
